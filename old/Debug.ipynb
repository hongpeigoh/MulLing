```python
from scipy.stats import friedmanchisquare
import scikit_posthocs

stat, p = friedmanchisquare(*data)
print('stat=%.3f, p=%.3f' % (stat, p))
if p > 0.05:
    print('Probably the same distribution')
else:
    print('Probably different distributions')
```
