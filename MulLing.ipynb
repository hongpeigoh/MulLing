### Tagged Eval

```python
import pandas as pd
df = pd.read_csv('dump/evalqueries.tsv', sep='\t')
print(df)
```

```python
models = ['baa','meta','bai','laser','metalaser','senlaser','senbai']
model_results = {}

for model in models:
    model_results[model] = {
        'enen': [],
        'enzh': [],
        'zhen': [],
        'zhzh': []
    }

for i  in range(400):
    q = df['query'][i]
    q_lang = df['query language'][i]
    to_lang = df['lang'][i]
    to_article = df['article index'][i]
    for model in models:
        results = query.mulling_annoy_query(my_object,
                                            q = q,
                                            model = model,
                                            k = 100,
                                            multilingual = False,
                                            lang_= q_lang,
                                            olangs = [to_lang])
        rresults = [result[1] for result in results]
        if to_article in rresults:
            model_results[model]['%s%s' % (q_lang, to_lang)].append(rresults.index(to_article))
        else:
            model_results[model]['%s%s' % (q_lang, to_lang)].append(-1)
    if (i+1)%20 == 0:
        print('%i queries done'%(len(model_results[model]['%s%s' % (q_lang, to_lang)])))
```

```python
def topk_acc(model, model_results, k, langlang):
    return len(list(filter(lambda x: x<=k and x!=-1, model_results[model][langlang])))/100

acc_results = []
for model in models:
    append_in = []
    append_in.append(model)
    for langlang in ['enen','enzh','zhen','zhzh']:
        for k in [1,5,10,50,100]:
            append_in.append(topk_acc(model, model_results, k, langlang))
    acc_results.append(append_in)
    
from pprint import pprint
pprint(acc_results)
```

```python
df2 = pd.DataFrame(acc_results)
pprint(df2)
```

```python
from matplotlib import pyplot as plt

models = ['baa','meta','bai','senbai','laser','metalaser','senlaser']
langlangs = ['enen','enzh','zhen','zhzh']
colors = ['r','b','g','brown']

plt.figure(figsize=(20,20))
for model in models:
    plt.subplot('42%i'% (models.index(model)+1))
    for langlang in langlangs:
        color = colors[langlangs.index(langlang)]
        accs = [topk_acc(model, model_results, k+1, langlang) for k in range(100)]
        fig = plt.plot(accs, color=color)
        plt.text(80, accs[-1]+0.01, langlang, color=color)

    plt.title('Results for %s model'% (model))
    plt.ylabel('MAP Scores for Top K')
    plt.xlabel('Result Number K')
    plt.ylim(0,1.1)

fig = plt.show()
```

```python
from math import floor, log
from annoy import AnnoyIndex
import pickle

a = pickle.load(open('pickle/metalasers_new.pkl', 'rb'))
for lang in a:
    t = AnnoyIndex(1024, 'angular')  # Length of item vector that will be indexed
    for i in range(len(a[lang])):
        t.add_item(i, a[lang][i][0])

    t.build(floor(log(len(a[lang])))) # 10 trees
    t.save('annoy/%s/%s1.ann' % (lang, 'metalaser'))
```

```python
import requests
import numpy as np
import json

address = '127.0.0.1:5050'

k = 20
q = 'high as boy'
model = 'metalaser'
lang = 'en'
multi = 'mono'

def get_results(q, k, model, lang, address, multi):
    url = 'http://' + address + '/query_' + multi
    params = {'q':q, 'model':model, 'lang':lang, 'k':k}
    resp = requests.get(url=url, params=params).json()
    return resp['results']

print('\n'.join(get_results(q=q, k=k, model=model, lang=lang, address=address, multi=multi)))
```
